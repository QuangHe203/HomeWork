# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv")
print(data.head())
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python

month = data['month_number']
total_profit = data['total_profit']
plt.plot(month, total_profit)
plt.xlabel('Month number')
plt.ylabel('Total per month')
plt.title('Company profit per month')
plt.xticks(range(1, 13))
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
month = data['month_number']
total_profit = data['total_profit']
line, = plt.plot(month, total_profit, marker='o', linestyle='--', color='red', markerfacecolor='black', label='profit data of last year')
plt.xlabel('Month Number')
plt.ylabel('Total per month')
plt.title('Company Sales data of last year')
plt.legend(handles=[line], loc='lower right')
plt.xticks(range(1, 13))
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
facecream_profit = data['facecream']
facewash_profit = data['facewash']
toothpaste_profit = data['toothpaste']
bathingsoap_profit = data['bathingsoap']
shampoo_profit = data['shampoo']
moisturizer_profit = data['moisturizer']
plt.plot(month, facecream_profit, marker='o', label='Face Cream')
plt.plot(month, facewash_profit, marker='o', label='Face Wash')
plt.plot(month, toothpaste_profit, marker='o', label='Toothpaste')
plt.plot(month, bathingsoap_profit, marker='o', label='Bathing Soap')
plt.plot(month, shampoo_profit, marker='o', label='Shampoo')
plt.plot(month, moisturizer_profit, marker='o', label='Moisturizer')
plt.xlabel('Thng')
plt.ylabel('Tng li nhun')
plt.title('Tng li nhun theo tng sn phm theo thng')
plt.legend(loc='upper left')
plt.xticks(range(1, 13))
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
month = data['month_number']
toothpaste_sales = data['toothpaste']
plt.scatter(month, toothpaste_sales, label = 'Toothpaste sales')
plt.xlabel('Month Number')
plt.ylabel('Number units Sold')
plt.title("Tooth paste Sales data")
plt.legend(loc='upper left')
plt.xticks(range(1, 13))
plt.grid(True, linestyle='--')
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
month = data['month_number']
face_cream_sales = data['facecream']
facewash_sales = data['facewash']
bar_width = 0.3
plt.bar(month - bar_width / 2, face_cream_sales, width=bar_width, label='Face Cream sales')
plt.bar(month + bar_width / 2, facewash_sales, width=bar_width, label='Facewash sales')
plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')
plt.title('Face Cream v Facewash Sales Data')
plt.grid(True, linestyle='--')
plt.legend(loc='upper left')
plt.xticks(range(1, 13))
plt.show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
month = data['month_number']
bathingsoap_sales = data['bathingsoap']
plt.xlabel('Month Number')
plt.ylabel("Sales units in number")
plt.title('bathingsoap sales data')
bar_width = 0.7
plt.bar(month, bathingsoap_sales, bar_width)
plt.grid(True, linestyle='--')
plt.xticks(range(1, 13))
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
plt.hist(data['total_profit'], bins=[150000, 200000, 250000, 300000], edgecolor='black', alpha=0.7)
plt.xticks(data['total_profit'])
plt.xlabel('Total Profit per Month in Dollar')
plt.ylabel('Frequency')
plt.title('Histogram of Most Common Profit Ranges')

plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
facecream_total = data['facecream'].sum()
facewash_total = data['facewash'].sum()
toothpaste_total = data['toothpaste'].sum()
bathingsoap_total = data['bathingsoap'].sum()
shampoo_total = data['shampoo'].sum()
moisturizer_total = data['moisturizer'].sum()
products = ['Facecream', 'Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
total_sales = [facecream_total, facewash_total, toothpaste_total, bathingsoap_total, shampoo_total, moisturizer_total]
plt.pie(total_sales, labels=products, autopct='%1.1f%%')
plt.axis('equal') 
plt.title('Sales data')
plt.legend(loc='lower right')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
plt.subplot(2, 1, 1)  # 2 dng, 1 ct, subplot 1
plt.plot(data['month_number'], data['bathingsoap'], marker='o', color='k')
plt.title('Sales data of Bathingsoap')
plt.subplot(2, 1, 2)
plt.plot(data['month_number'], data['facewash'], marker='o', color='r')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.xticks(range(1, 13))
plt.title('Sales data of facewash')
plt.subplots_adjust(hspace=0.5)
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
plt.stackplot(data['month_number'], data['facecream'], data['facewash'], data['toothpaste'], data['bathingsoap'], data['shampoo'], data['moisturizer'], labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])

plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
