### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('titanic_train.csv')
titanic.head(10)
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId', inplace=True)
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
male_count = (titanic['Sex'] == 'male').sum()
female_count = (titanic['Sex'] == 'female').sum()
labels = ['Male', 'Female']
sizes = [male_count, female_count]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal') 
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
import matplotlib.pyplot as plt

colors = {'male': 'blue', 'female': 'red'}
fig, ax = plt.subplots()
for gender in titanic['Sex'].unique():
    titanic_gender = titanic[titanic['Sex'] == gender]
    ax.scatter(titanic_gender['Age'], titanic_gender['Fare'], c=colors[gender], label=gender, alpha=0.5)

ax.set_title('Fare paid vs Age by Gender')
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.legend()
plt.show()
```

### Step 7. How many people survived?

```python
survived_num = (titanic['Survived'] == 1).sum()
print("Number of survived people: ", survived_num)
```

### Step 8. Create a histogram with the Fare payed

```python
Fare = titanic['Fare'].tolist()
profit_range = [20, 40, 60, 80, 100]
plt.hist(Fare ,profit_range)

plt.xlabel('Fare')
plt.ylabel('Actual Profit in dollar')
plt.show()
```
