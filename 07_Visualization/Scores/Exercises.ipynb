### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
        'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
        'age': [42, 52, 36, 24, 73],
        'female': [0, 1, 1, 0, 1],
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data)
print(df)
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'])
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.title('Scatterplot of preTestScore vs postTestScore (Size by Age)')
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(df['preTestScore'], df['postTestScore'], s=4.5*df['postTestScore'], c=df['female'])
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')
plt.title('Scatterplot of preTestScore vs postTestScore (Color by sex)')
plt.show()
```

### BONUS: Create your own question and answer it.

Question: How does the distribution of post-test scores compare between males and females in a given dataset?


```python
male_df = df[df['female'] == 0]
female_df = df[df['female'] == 1]
plt.hist(male_df['postTestScore'], alpha=0.5, label='Male')
plt.hist(female_df['postTestScore'], alpha=0.5, label='Female')
plt.xlabel('Post-Test Score')
plt.ylabel('Frequency')
plt.title('Distribution of Pre-test Scores by Gender')
plt.legend()
plt.show()
```
