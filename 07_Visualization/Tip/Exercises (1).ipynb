### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
import urllib.request

data = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv' )   
```

### Step 3. Assign it to a variable called tips

```python
tips = data
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop('Unnamed: 0', axis=1, inplace=True)
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(tips["total_bill"])
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x="tip", y="day", hue="sex", data=tips)
plt.title("Scatter Plot of Tip vs. Day by Sex")
plt.xlabel("Tip")
plt.ylabel("Day")
plt.legend(title="Sex")
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips)
plt.title("Box Plot of Total Bill by Day and Time")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.legend(title="Time")
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
dinner_tips = tips[tips["time"] == "Dinner"]["tip"]
lunch_tips = tips[tips["time"] == "Lunch"]["tip"]
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.histplot(dinner_tips, kde=False)
plt.title("Histogram of Tip Values for Dinner")
plt.xlabel("Tip")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
sns.histplot(lunch_tips, kde=False)
plt.title("Histogram of Tip Values for Lunch")
plt.xlabel("Tip")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips[tips["sex"] == "Male"])
plt.title("Scatter Plot of Total Bill vs. Tip for Males")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend(title="Smoker")

plt.subplot(1, 2, 2)
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips[tips["sex"] == "Female"])
plt.title("Scatter Plot of Total Bill vs. Tip for Females")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.legend(title="Smoker")

plt.tight_layout()
plt.show()
```

### BONUS: Create your own question and answer it using a graph.

Question: What is the distribution of total bill values for smokers and non-smoker

```python
smoker_total_bills = tips[tips["smoker"] == "Yes"]["total_bill"]
non_smoker_total_bills = tips[tips["smoker"] == "No"]["total_bill"]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.histplot(smoker_total_bills, kde=False)
plt.title("Histogram of Total Bill Values for Smokers")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
sns.histplot(non_smoker_total_bills, kde=False)
plt.title("Histogram of Total Bill Values for Non-smokers")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
